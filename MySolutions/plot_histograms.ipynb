```python
fig,ax=plt.subplots()

ax=df1[2000].hist()
ax.set_xlim(0,100)
ax.set_xlabel("Life expectancy [years]")
ax.set_ylabel("number of countries")

plt.savefig('histo.png')
```

```python
year=2000
for bin in [5,10,20]:
    fig,ax=plt.subplots()

    ax=df1[year].hist(bins=bin)
    #ax.set_xlim(0,100)
    #ax.set_ylim(0,80)
    plt.axis([0.0, 100.0, 0.0, 80.0])
    ax.set_xlabel("Life expectancy [years]")
    ax.set_ylabel("number of countries")

    plt.savefig('histo_bins_'+str(bin)+'.png')
    plt.title('Life Expendency in '+str(year)+' using Bins of '+str(bin)+' years')
```
