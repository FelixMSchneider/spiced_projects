```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_excel("../../garlic-boosting-encounter-notes/01_week/data/gapminder_lifeexpectancy.xlsx")
```

```python
df1=df1.dropna()
```

```python
meanLE=[]
for year in list(df1.columns[1:]):
    meanLE.append((df1[year].mean()))
medianLE=[]
for year in list(df1.columns[1:]):
    medianLE.append((df1[year].median()))
    
plt.plot(df1.columns[1:], meanLE)    
plt.plot(df1.columns[1:], medianLE)  
plt.xlabel("year")
plt.ylabel("global mean life expectancy");

plt.xlim(1800,2020)
```
